```python
import numpy as np
```

```python

data = np.genfromtxt('StressLevelDataset.csv', delimiter=',', skip_header=1)
data
```

Визначимо стандартне відхилення для стовпрців

```python
std_deviation = np.std(data, axis=0)
print("Стандартне відхилення:", std_deviation)
```

Поглянемо на кореляцію між обсягом навчання та рівнем депресії

```python
study_load = data[:, 13]  
depression = data[:, 3]  

correl = np.corrcoef(study_load, depression)[0, 1]

print("Кореляція між обсягом навчання та депресією:", correl)
```

Візуалізуємо це

```python
import matplotlib.pyplot as plt

study_load = data[:, 13]  
depression = data[:, 3]  

plt.figure(figsize=(10, 6))
plt.scatter(study_load, depression, alpha=0.5, c='blue', edgecolors='none', s=30)

plt.title('Взаємозв\'язок між обсягом навчання та депресією')
plt.xlabel('Обсяг навчання')
plt.ylabel('Рівень депресії')
plt.grid(True)
plt.show()
```
