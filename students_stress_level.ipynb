```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("StressLevelDataset.csv")
data
```

Визначте фактори стресу судентів:

```python
#сумуємо усі стовбці
sums = data.sum()
#назви фаторів
stress_factors = data.columns

#будуємо діаграму
plt.figure(figsize=(8, 10))
plt.pie(sums, labels=stress_factors, autopct='%1.1f%%')
plt.title('Stress factors')
plt.show()
```

Зобразіть середні рівні стресу для студентів, які мають позакласні заходи, або без них:

```python
# Розділимо дані на дві групи
with_extracurricular = data[data['extracurricular_activities'] == 1]['stress_level']
without_extracurricular = data[data['extracurricular_activities'] == 0]['stress_level']

# Обчислимо середніх значень для кожної групи
mean_with_extracurricular = with_extracurricular.mean()
mean_without_extracurricular = without_extracurricular.mean()

#Побудуємо графік
plt.bar(['With extracurricular activities', 'Without extracurricular activities'], [mean_with_extracurricular, mean_without_extracurricular])
plt.title('Average stress levels for students with and without extracurricular activities')
plt.ylabel('Average stress level')
plt.show()
```

Який відсоток студентів повідомляють про тривогу вище середнього рівня?

```python
# Розрахуємо середньє значення тривоги 
anxiety = data["anxiety_level"].mean()
# Підрахуємо та визначмо к-ть студентів у кожній категорії тривоги
anxiety_levels = data["anxiety_level"].value_counts()
high_anxiety = anxiety_levels[anxiety_levels.index > anxiety].sum()
# Порахуємо відсоток студентів з тривогою вище середнього рівня
high_anxiety_percentage = (high_anxiety / len(data)) * 100

#Результати
print(f'Середнє значення тривоги: {anxiety:.2f}')
print(f'Відсоток студентів з тривогою вище середнього рівня: {high_anxiety_percentage:.2f}%')
```
